```python
# Before we proceed, we will first see how we can handle pdf file in python.
from pypdf import PdfReader

```

```python
# pip install pypdf
```

```python
pdf_file = PdfReader("Skill Match.pdf")
```

```python
for page in pdf_file.pages:
    print(page.extract_text())
```

```python
content = ''
for page in pdf_file.pages:
    content = content + page.extract_text() + '\n'
```

```python
print(content)
```
